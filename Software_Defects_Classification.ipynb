# **Seperating Dependent & Independent Variables**

---



```python
X = df.drop(["id","defects"], axis=1)
y = df["defects"]
```

```python
X.head()
```

```python
y.head()
```

```python
X.shape, y.shape
```

# **Encoding**

---



```python
d_map = {False:0, True:1}
y = y.map(d_map)
```

```python
y.head()
```

# **Splitting into Training & Testing Set**

---



```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,
                                                    random_state=0)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# **Model Building - Logistic Regression**

---



```python
from sklearn.linear_model import LogisticRegression
lr_model = LogisticRegression()
```

```python
lr_model.fit(X_train, y_train)
```

```python
lr_model.score(X_test, y_test)
```

```python
y_preds1 = lr_model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
print(confusion_matrix(y_test, y_preds1))
```

# **Model Building - Decision Trees Classifier**

---



```python
from sklearn.tree import DecisionTreeClassifier
dtc_model = DecisionTreeClassifier()
```

```python
dtc_model.fit(X_train, y_train)
```

```python
dtc_model.score(X_test, y_test)
```

```python
y_preds2 = dtc_model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
print(confusion_matrix(y_test, y_preds2))
```

# **Model Building - XGB Classifier**

---



```python
from xgboost import XGBClassifier
xgbc_model = XGBClassifier()
```

```python
xgbc_model.fit(X_train, y_train)
```

```python
xgbc_model.score(X_test, y_test)
```

```python
y_preds3 = xgbc_model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
print(confusion_matrix(y_test, y_preds3))
```
